Implement the XOR and (¬XOR) logic operation using a multi-perceptron network, and analyze how multiple perceptrons overcome the limitations of a single perceptron

```python
import numpy as np

# Step activation function
def step(x):
    return 1 if x >= 0 else 0

# XOR network
def xor_network(x1, x2):
    # Hidden Layer

    # H1 = OR gate
    h1 = step((1 * x1) + (1 * x2) - 0.5)

    # H2 = NAND gate
    h2 = step((-1 * x1) + (-1 * x2) + 1.5)

    # Output Layer (AND gate)
    y = step((1 * h1) + (1 * h2) - 1.5)

    return y

# Testing XOR truth table
print("XOR Truth Table")
print("x1 x2 | XOR")
for x1 in [0, 1]:
    for x2 in [0, 1]:
        print(f"{x1}  {x2}  |  {xor_network(x1, x2)}")
```

Demonstrate that the thresholding logic used by perceptron is very harsh.


```python
def step(z):
    return 1 if z >= 0 else 0

def perceptron(x1, x2):
    z = x1 + x2 - 1
    return z, step(z)

# Test values near boundary
test_values = [(0.49, 0.49), (0.50, 0.50), (0.51, 0.51)]

for x1, x2 in test_values:
    z, y = perceptron(x1, x2)
    print(f"x1={x1}, x2={x2}, z={z:.2f}, output={y}")
```

Implement the Perceptron Learning Algorithm and study the effect of weight updates on convergence for a binary decision problem such as determining whether a user would like to watch a movie.
Note: Consider a small dataset(design your own excel csv sheet) of movie records with Boolean or real-valued features, for example:
 f1​: Is actor Matt Damon present
 f2​: Is the genre Thriller
 f3​: Is the director Christopher Nolan
 f4: IMDb rating (scaled between 0 and 1)
The output label represents like (1) or dislike (0). Train the perceptron model using these features and observe how weight updates influence convergence and classification performance.
Test with a sample record to show whether a perceptron properly classifies it or not.


```python
import numpy as np

# Dataset
X = np.array([
    [1,1,1,0.90],
    [1,0,1,0.80],
    [0,1,1,0.70],
    [0,1,0,0.60],
    [1,0,0,0.40],
    [0,0,0,0.30],
    [0,1,0,0.20],
    [1,0,0,0.20]
])

y = np.array([1,1,1,1,0,0,0,0])

# Step activation
def step(z):
    return 1 if z >= 0 else 0

# Training Perceptron
def train_perceptron(X, y, lr=0.1, epochs=20, use_bias=True):
    n_samples, n_features = X.shape
    w = np.zeros(n_features)
    b = 0

    for epoch in range(epochs):
        errors = 0
        for i in range(n_samples):
            z = np.dot(X[i], w) + (b if use_bias else 0)
            y_pred = step(z)

            update = lr * (y[i] - y_pred)

            if update != 0:
                w += update * X[i]
                if use_bias:
                    b += update
                errors += 1

        print(f"Epoch {epoch+1}, Errors: {errors}, Weights: {w}, Bias: {b}")

        if errors == 0:
            break

    return w, b
```

i) Check with MP Perceptron  (without weights and bias)

```python
def mp_perceptron(x):
    return step(np.sum(x))

# Test sample
sample = np.array([0,1,1,0.85])
print("MP Output:", mp_perceptron(sample))
```

ii) Check with Perceptron (with weights)

```python
w_no_bias, b_no_bias = train_perceptron(X, y, use_bias=False)

z = np.dot(sample, w_no_bias)   # no bias added
prediction_no_bias = step(z)

print("Perceptron (No Bias) Prediction:", prediction_no_bias)
```

```python
w_bias, b_bias = train_perceptron(X, y, use_bias=True)

z = np.dot(sample, w_bias) + b_bias
prediction_bias = step(z)

print("Perceptron (With Bias) Prediction:", prediction_bias)
```

How many Boolean functions can be designed using three binary inputs?


```python
import numpy as np
from itertools import product

# --------------------------------------------------
# Step activation
# --------------------------------------------------
def step(z):
    return 1 if z >= 0 else 0

# --------------------------------------------------
# Train perceptron
# --------------------------------------------------
def train_perceptron(X, y, lr=0.1, epochs=50):
    w = np.zeros(X.shape[1])
    b = 0

    for _ in range(epochs):
        errors = 0
        for i in range(len(X)):
            y_pred = step(np.dot(w, X[i]) + b)
            error = y[i] - y_pred
            w += lr * error * X[i]
            b += lr * error
            errors += abs(error)
        if errors == 0:
            return True  # converged
    return False  # not converged

# --------------------------------------------------
# Inputs for 3-bit Boolean functions
# --------------------------------------------------
X = np.array(list(product([0, 1], repeat=3)))
# X has 8 rows now

# --------------------------------------------------
# Generate all Boolean functions (3 inputs → 8 outputs)
# --------------------------------------------------
boolean_functions = list(product([0, 1], repeat=8))

linearly_separable = 0
not_separable = 0
failed_functions = []

for func in boolean_functions:
    y = np.array(func)
    converged = train_perceptron(X, y)

    if converged:
        linearly_separable += 1
    else:
        not_separable += 1
        failed_functions.append(func)

# --------------------------------------------------
# Results
# --------------------------------------------------
print("Total Boolean functions (n=3):", len(boolean_functions))
print("Linearly separable:", linearly_separable)
print("Not linearly separable:", not_separable)
```

Design and implement a Multi-Layer Perceptron (MLP) architecture capable of realizing all basic Boolean functions  

Hint:

By finding the weights w1,w2,w3,w4 such that the truth table of any boolean function can be represented by this network.
Also Each perceptron in the middle layer fires only for a specific input (and no two perceptrons fire for the same input)


```python
import numpy as np
from itertools import product

# Input combinations for 2 variables
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

# Step function
def step(x):
    return 1 if x >= 0 else 0

# Train perceptron
def train_perceptron(X, y, epochs=20, lr=0.1):
    w = np.zeros(X.shape[1])
    b = 0

    for _ in range(epochs):
        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            pred = step(z)
            error = y[i] - pred

            w += lr * error * X[i]
            b += lr * error

    # Check if perfectly learned
    for i in range(len(X)):
        if step(np.dot(X[i], w) + b) != y[i]:
            return False
    return True


# Generate all 16 Boolean functions
functions = list(product([0,1], repeat=4))

linearly_separable = []
non_separable = []

for idx, f in enumerate(functions):
    y = np.array(f)
    if train_perceptron(X, y):
        linearly_separable.append((idx+1, f))
    else:
        non_separable.append((idx+1, f))


# Results
print("Linearly Separable Functions:", len(linearly_separable))
for i, f in linearly_separable:
    print(f"f{i}: {f}")

print("\nNon-Linearly Separable Functions:", len(non_separable))
for i, f in non_separable:
    print(f"f{i}: {f}")
```
